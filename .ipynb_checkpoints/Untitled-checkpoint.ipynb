```python
def prep(data):
    cols= ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','target']
    data.columns = cols
    x_cols = ['workclass','education','marital_status','occupation','relationship','race','sex','native_country','target']
    label_encoder = preprocessing.LabelEncoder()
    for i in x_cols:
        data[i]= label_encoder.fit_transform(data[i]) 
    data = data.drop(['fnlwgt'], axis = 1)
    return data
```

```python
def split(df):
    X=df.drop(['target'],axis=1)
    y=df.target
    X, y = oversample.fit_resample(X, y)
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8,random_state=1)
    print("Input Training:",X_train.shape)
    print("Input Test:",X_test.shape)
    print("Output Training:",y_train.shape)
    print("Output Test:",y_test.shape)
    return X_train, X_test, y_train, y_test
```

```python
#logistic regression
def lr(X_train, X_test, y_train, y_test):
    from sklearn.linear_model import LogisticRegression
    from sklearn import metrics
    logreg = LogisticRegression()
    logreg.fit(X_train, y_train)
    y_pred = logreg.predict(X_test)
    print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
    return y_pred, logreg
```

```python
#confusion matrix
def cm(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:\n",cm)
    print("Classification Report:\n",classification_report(y_test, y_pred))
```

```python
#roc auc plot
def auc_plot(X_train, y_train, X_test, y_test):
    rt=RandomForestClassifier(n_estimators=9,min_samples_leaf= 2, max_features= 2, max_depth= 10 )
    rt.fit(X_train,y_train)
    logit_roc_auc = roc_auc_score(y_test, rt.predict(X_test))
    fpr, tpr, thresholds = roc_curve(y_test, rt.predict_proba(X_test)[:,1])
    plt.figure()
    plt.plot(fpr, tpr, label='Random Forest Classifier (area = %0.2f)' % logit_roc_auc)
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([-0.01, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

```python
#decision tree
def dt(X_train, X_test, y_train, y_test):
    dt = DecisionTreeClassifier(criterion="gini", max_depth=3)
    dt = dt.fit(X_train,y_train)
    y_pred = dt.predict(X_test)
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    return y_pred
```

```python
#random forest
def rf(X_train, X_test, y_train, y_test):
    scores = []
    for i in range(10,100,2):
        rt=RandomForestClassifier(n_estimators=10)
        rt.fit(X_train,y_train)
        y_pred=rt.predict(X_test)
        scores.append(metrics.accuracy_score(y_test, y_pred))
    
#     print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    return y_pred, rf, scores
```
